```python
#ML Image Dataset Creator

import cv2
import cv2 as cv
import numpy as np
import os

webCam = cv2.VideoCapture(0)

cv.namedWindow('Crop')

print(webCam.isOpened())
print(webCam.get(cv2.CAP_PROP_FRAME_WIDTH))
print(webCam.get(cv2.CAP_PROP_FRAME_HEIGHT))

save = 0
iterator = 0
letter = 'Blank' #sub folder name
path = 'C:/Users/Stefan/Desktop/MLDB/' + letter + '/'
fileName = path + letter + str(iterator) + '.png'

while(webCam.isOpened()):
    
    capRet, camFeed = webCam.read()
    if capRet == True:
        
        camFeed = cv2.rectangle(camFeed, (225, 145), (415, 335), (0, 255, 0), 0)
        
        croppedImage = camFeed[145:335, 225:415]
        croppedImage = cv2.cvtColor(croppedImage, cv2.COLOR_BGR2GRAY)
        _, croppedImage = cv2.threshold(croppedImage, 127, 255, cv.THRESH_BINARY)
        
        if cv2.waitKey(1) == 32:
            save = 1
            if not os.path.isdir(path):
                os.makedirs(path)
        elif cv2.waitKey(1) > 0:
            break
        
        cv2.imshow('WebCam Output', camFeed)
        cv2.imshow('Crop', croppedImage)
        if save == 1:
            cv2.imwrite(fileName, croppedImage)
            iterator = iterator + 1
            fileName = path + letter + str(iterator) + '.png'
        if iterator == 1000: #stop condition
            break
        
    else:
        break
        
webCam.release()
cv2.destroyAllWindows()
```
